#: Reshaping NumPy Arrays

```python
#Create a 1D NumPy array, e.g., arr_1d = np.arange(12).
oneDArray = np.arange(12)
oneDArray
```

```python
#Reshape the 1D array into a 2D array with 3 rows and 4 columns
twoDArray = oneDArray.copy().reshape((3,4))
twoDArray
```

```python
#Reshape the 2D array into a 3D array with 2 matrices, each having 3 rows and 2 columns.
threeDArray = twoDArray.copy().reshape((2,3,2))
threeDArray
```

#Visualizing Reshaped Arrays

```python
#Use Matplotlib to visualize the reshaped arrays.
plt.plot(oneDArray)
plt.title("One Dimensional Array")
plt.show()
```

```python
plt.boxplot(oneDArray)
plt.title("One Dimesional Array")
plt.show()
```

```python
x = twoDArray[0,:]
y = twoDArray[1,:]
z = twoDArray[2,:]
ax=plt.axes(projection="3d")
ax.scatter(x,y,z, s = 400)
ax.set_title("3D Array")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.contourf(twoDArray, cmap = "viridis")
plt.title('2D Array')
plt.colorbar()
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.grid(True)
plt.show()
```

```python
ax = plt.axes(projection="3d")
X, Y =np.meshgrid(threeDArray, threeDArray)
Z = X*Y
ax.scatter(X,Y,Z)
ax.set_title("3D Array")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
plt.show()
```

```python
ax = plt.axes(projection="3d")
X, Y =np.meshgrid(threeDArray[0,:], threeDArray[1,:])
Z = Y * X
ax.scatter(X,Y,Z, s = 150)
ax.set_title("3D Array")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
plt.show()
```

```python
#Create a subplot for each array.
x = twoDArray[0,:]
y = twoDArray[1,:]
z = twoDArray[2]
fig,axs = plt.subplots(3)
fig.suptitle('2D subplots')
axs[0].plot(x,y)
axs[1].plot(x,-y)
axs[2].plot(x,z)
```

```python
x = threeDArray[0,:]
y = threeDArray[1,:]
fig,axs = plt.subplots(2)
fig.suptitle('3D subplots')
axs[0].plot(x,y)
axs[1].plot(x,-y)
```

```python
#Use imshow() to display the values in a matrix.
#Add appropriate titles and labels.

plt.imshow(twoDArray,interpolation='nearest')
plt.colorbar()
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('2D Array')
plt.show()
```

```python
plt.imshow(threeDArray[:,:,0], interpolation='nearest')
plt.colorbar()
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('3D Array')
plt.show()
```

#Advanced Reshaping - Transposing

```python
#Create a 2D array, e.g., arr_2d = np.random.rand(3, 4)
arr2d = np.random.rand(3,4)
arr2d
```

```python
#Transpose the array using np.transpose() or the .T attribute.
a_t = arr2d.T
a_t
```

```python
#Visualize the original and transposed arrays using Matplotlib.

plt.hist(arr2d, bins = 3)
plt.title("Untransposed 2D array")
plt.ylabel('Frequency')
plt.xlabel('Values')
plt.show()
```

```python
plt.hist(a_t, bins = 3)
plt.title("Transposed 2D array")
plt.ylabel('Frequency')
plt.xlabel('Values')
plt.show()
```
